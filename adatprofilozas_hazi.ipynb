## **Adatprofilozás**

### 1\. DVD-k nettóárának minimális, maximális, átlagos értéke és szórása.

```sql
SELECT min(nettoar) As 'Minimum nettóár',
       max(nettoar) As 'Maximum nettóár',
       round(avg(nettoar),2) As ' Átlag',
       round(stdev(nettoar),2) As 'Szórás'

FROM dvd
WHERE nettoar IS NOT NULL
```

### 2\. DVD-k átlagos nettó ára stílusonként, ár szerint csökkenő sorrendben.

```sql
SELECT stilus,
       avg(nettoar) 'átlagos nettó ár'
FROM dvd
GROUP BY stilus
Having stilus is not NULL
ORDER BY avg(nettoar) DESC
```

### 3\. Egyedi értékek száma

```sql
SELECT COUNT(DISTINCT nettoar) As 'egyedi értékek száma',
       COUNT(*) As 'összes'
FROM dvd
WHERE nettoar IS NOT NULL
```

### 4\. Mennyi DVD ára tér el nagyon az átlagtól(4682,34)

```sql
SELECT IIF(abs(nettoar-4682.34)<=400, 'Nagy', 'Kicsi') As 'Átlagtól való eltérés',
count(*) As 'darabszám'
FROM dvd 
GROUP by IIF(abs(nettoar-4682.34)<=400, 'Nagy', 'Kicsi')

```

### 5\. Különböző DVD-k száma stílusonként, db szám szerint csökkenő sorrendben

```sql
Select stilus,
       count(*) As 'darabszám'
FROm dvd
GROUP BY stilus
HAVING stilus is not null
order by count(*) DESC
```

### 6\. NULL értékek száma a stílusok között

```sql
SELECT CASE WHEN stilus IS NULl THEN 'Null'
ELSE 'Not Null' END As 'Null/Nem null',
count(*) As 'darabszám'
FROM dvd
GROUP BY CASE WHEN stilus IS NULl THEN 'Null'
ELSE 'Not Null'END
```

### 7\. Karakterlánc hossza( minimum, maximum, átlag)

```sql
SELECT min(len(stilus)) AS 'legrövidebb hossz',
       max(len(stilus)) As 'leghosszabb hossz',
       avg(len(stilus)) As 'átlagos hossz'
FROM dvd
WHERE stilus IS NOT NULL
```

### 8\. Első és utolsó karakterlánc abc sorrendbe nézve

```sql
SELECT min(stilus) As 'első',
       max(stilus) As 'utolsó'
FROM dvd
WHERE stilus IS NOT NULL
AND len(stilus)>0
```
